```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

```python
set(df['Name']) #เลือกค่าที่ไม่ซ้ำ 
```

```python
df['Name'].unique() #เลือกค่าที่ไม่ซ้ำ 
```

##Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name') #df=ชื่อตาราง Name=จัดกลุ่มตามname
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength'])
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength,'SepalWidth','PetalWidth'])
```


##change color

```python
from matplotlib import pyplot as plt
```

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize']=[25,25] #กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest') #สีสว่าง ค่ามาก สีเข้ม ค่าน้อย
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap = 'hot') #cmap กำหนดสี
```

#Subplot

```python
matplotlib.rcParams['figure.figsize']=[10,10] #ขนาด
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest') #50 = ตัวที่50

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest') # 50-100

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest')# 100-150
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',alpha = 0.2)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',alpha = 0.8) #alpha คือความโปร่งใส 1-0 
```

```python
plt.subplot(1,2,1)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5,0.5],[0,0,0.8,0.5],[0.2,0.4,0.7,0.5]]) # ตัวเลขที่สี่คือ alpha

plt.subplot(1,2,2)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
```
