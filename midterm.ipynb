1.หาพื้นที่ (district_name)ที่มีจำนวนโครงการเยอะที่สุดและมีความต้องการเช่าเยอะที่สุด (5 คะแนน)

```python
dataall = project.merge(rent, left_on='row_number', right_on='row_number')
dataall.head()
```

```python
dataall2.groupby("district_name_th_y")[["name_th"]].max().sort_values(by=["name_th"], ascending=False)#_y = ชื่อพื้นที่ทั่วประเทศ ทีมีโครงการเยอะสุด
ans_1_1 = dataall2.groupby("district_name_th_y")[["name_th"]].max().reset_index().sort_values(by=["name_th"], ascending=False)
ans_1_1
```

```python
dataall2.groupby("district_name_th_y")[["volume_of_rental"]].max().sort_values(by=["volume_of_rental"], ascending=False) #_y = ชื่อพื้นที่ที่มีความต้องการเช่ามากสุด
ans_1_2 = dataall2.groupby("district_name_th_y")[["volume_of_rental"]].max().reset_index().sort_values(by=["volume_of_rental"], ascending=False)
ans_1_2
```

5. (เฉพาะ กรุงเทพมหานคร) เปรียบเทียบราคาโครงการที่แพงที่สุดในย่าน (subdistrict)ที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาสูงที่สุด (eating_nighttime_price) กับ ย่าน (subdistrict)ที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาต่ำที่สุด (eating_nighttime_price) (10 คะแนน)

```python
project_2 = project[['row_number','name_th','price_min']]
eat_2 = eat[['row_number','subdistrict_name_th','province_name_th','eating_nighttime_price']]
```

```python
Dataall5= project_2.merge(eat_2, left_on='row_number', right_on='row_number')
Dataall5
```

```python
all5_BKK =Dataall5[Dataall5['province_name_th']=='กรุงเทพมหานคร']
all5_BKK
```

```python
BKK = all5_BKK.groupby('subdistrict_name_th')[['eating_nighttime_price','name_th','price_min','subdistrict_name_th']].max()
BKK.sort_values('eating_nighttime_price',ascending=0)
#ย่านที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาสูงที่สุด คือ ลำต้อยติ่ง เเละโครงการที่เเพงที่สุดในย่านนั้นคือ วิลล่า สราสินี รีสอร์ท แอท โฮม 
```

```python
BKK = all5_BKK.groupby('subdistrict_name_th')[['eating_nighttime_price','name_th','price_min','subdistrict_name_th']].max()
BKK.sort_values('eating_nighttime_price',ascending=1)
#ย่านที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาต่ำที่สุด คือ ลุมพินี เเละโครงการที่เเพงที่สุดในย่านนั้นคือ บ้านกลางเมือง สวิสทาวน์ 
```
