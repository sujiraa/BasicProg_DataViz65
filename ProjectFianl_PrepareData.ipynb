```python
#จำนวนห้องนอน
#count_room_bed 
count_room_bedclean = table4[['count_room_bed']].fillna(1)
```

```python
#count_room_bath 
#จำนวนห้องน้ำ
count_room_bathclean = table4[['count_room_bath']].fillna(1)
```
